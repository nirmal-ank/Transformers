```python
from transformers import pipeline
```

```python
financial_bert = pipeline('text-classification', model='ProsusAI/finbert', tokenizer='ProsusAI/finbert')
```

```python
financial_bert("Amazon earnings beat expectations")
```

```python
financial_bert("The stock price was normal")
```

```python
financial_bert("The stock price fell down")
```

```python
financial_bert.model
```

```python
from transformers import BertForQuestionAnswering
```

```python
bert_qa = BertForQuestionAnswering.from_pretrained('bert-base-uncased')
bert_qa
```

```python
model_name = "timpal0l/mdeberta-v3-base-squad2"
qa = pipeline(model = model_name , tokenizer = model_name, task = "question-answering")
```

```python
question = "What is the capital of Nepal?"
context = "The capital of Nepal is Kathmandu"
qa(question = question, context = context)
```

```python
qa.model
```
