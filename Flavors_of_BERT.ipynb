```python
from transformers import pipeline
```

```python
nlp = pipeline("fill-mask", model="bert-base-cased")
```

```python
print(type(nlp.model))
```

```python
preds = nlp(f"Honesty is the best {nlp.tokenizer.mask_token}.")

print('Honesty is the best ***.')

for p in preds:
    print(f"Token:{p['token_str']}. Score: {100*p['score']:,.2f}%")
```

```python
nlp2 = pipeline("fill-mask", model='roberta-base')

print(type(nlp2.model))

preds = nlp2(f"If you don’t {nlp2.tokenizer.mask_token} at the sign, you will get a ticket")

print('If you don’t *** at the sign, you will get a ticket')

for p in preds:
    print(f"Token:{p['token_str']}. Score: {100*p['score']:,.2f}%")
```

```python
nlp3 = pipeline("fill-mask", model='distilroberta-base')

print(type(nlp3.model))

preds = nlp3(f"If you don’t {nlp3.tokenizer.mask_token} at the sign, you will get a ticket")

print('If you don’t *** at the sign, you will get a ticket')

for p in preds:
    print(f"Token:{p['token_str']}. Score: {100*p['score']:,.2f}%")
```

```python
nlp4 = pipeline("fill-mask", model='distilbert-base-cased')  # Using a flavor of BERT called DistilBERT

print(type(nlp4.model))

preds = nlp4(f"If you don’t {nlp4.tokenizer.mask_token} at the sign, you will get a ticket")

print('If you don’t *** at the sign, you will get a ticket')

for p in preds:
    print(f"Token:{p['token_str']}. Score: {100*p['score']:,.2f}%")
```
